```python
# Step-12. View the distribtution of the various categorical data/columns and plot each one of them.

# 1. Loan Status Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Loan_Status', data=df)
plt.title("Loan Status Distribution yes-0,No-1")
plt.show()
```

### Around 70% of loan applications were approved, while 30% were rejected. This indicates the dataset is slightly imbalanced, and approved cases dominate.

```python
# 2. Gender vs Loan Status
plt.figure(figsize=(6,5))
sns.countplot(x='Gender', hue='Loan_Status', data=df, palette='Set2')

plt.title("Male(0)/Female(1) Vs Loan Approved")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Loan Status")
plt.show()
```

### Male applicants have a higher loan approval rate compared to females, which could be due to higher income levels among male applicants in the dataset.

```python
# 3. Marriage Status vs Loan Status
plt.figure(figsize=(6,5))
sns.countplot(x='Married', hue='Loan_Status', data=df, palette='Set2')

plt.title("Married(0)/Unmarried(1) Vs Loan Approved")
plt.xlabel("Married")
plt.ylabel("Count")
plt.legend(title="Loan Status")
plt.show()
```

### Married applicants tend to get more loans approved, possibly because combined family income makes them more eligible.

```python
# 4. Education vs Loan Status
plt.figure(figsize=(6,5))
sns.countplot(x='Education', hue='Loan_Status', data=df, palette='Set2')

plt.title("Graduate(0)/Non-Graduates(1) Vs Loan Approved")
plt.xlabel("Education")
plt.ylabel("Count")
plt.legend(title="Loan Status")
plt.show()
```

### Graduates had higher approval rates than non-graduates, showing that education level may play a role in loan eligibility.

```python
# 5. Self_Employed vs Loan Status
plt.figure(figsize=(6,5))
sns.countplot(x='Self_Employed', hue='Loan_Status', data=df, palette='Set2')

plt.title("Self Employed(0)/Salaried applicants(1) Vs Loan Approved")
plt.xlabel("Self_Employed")
plt.ylabel("Count")
plt.legend(title="Loan Status")
plt.show()
```

### Self-Employed applicants actually had a slightly higher loan approval rate compared to salaried applicants.

```python
# 6. Property Area  vs Loan Status
plt.figure(figsize=(6,5))
sns.countplot(x='Property_Area', hue='Loan_Status', data=df, palette='Set2')

plt.title("Property Area(0=Semiurban, 1=Urban, 2=Rural) Vs Loan Approved")
plt.xlabel("Property_Area")
plt.ylabel("Count")
plt.legend(title="Loan Status")
plt.show()
```

### Semiurban applicants had the highest approval rates, followed by urban, while rural areas had the lowest approvals

```python
# 7. Income and Loan Amount Distribution
plt.scatter(df['ApplicantIncome'], df['LoanAmount'], c=df['Loan_Status'], cmap='bwr')
plt.title("Applicant Income vs Loan Amount")
plt.xlabel("Applicant Income")
plt.ylabel("Loan Amount")
plt.show()
```

### Most applicants had moderate incomes, with a few outliers earning very high incomes. Loan amounts were typically small to medium, with fewer large loans.

```python
# 8. Boxplots (Income/Loan vs Loan_Status)
# List of numerical features to compare
num_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

for col in num_cols:
    plt.figure(figsize=(7,5))
    sns.boxplot(x='Loan_Status', y=col, data=df, palette='Set2')
    plt.title(f"{col} vs Loan Status")
    plt.xlabel("Loan Status (0=Not Approved, 1=Approved)")
    plt.ylabel(col)
    plt.show()
```


- **ApplicantIncome vs Loan_Status** → Higher applicant income alone does not guarantee loan approval.  
- **CoapplicantIncome vs Loan_Status** → Coapplicant income has less impact on loan approval decisions.  
- **LoanAmount vs Loan_Status** → Moderate loan amounts are more likely to get approved than very high ones.  


```python
# 9. Scatterplot (Applicant Income vs Loan Amount)
plt.figure(figsize=(7,5))
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', hue='Loan_Status', data=df)
plt.title("Applicant Income vs Loan Amount")
plt.xlabel("Applicant Income")
plt.ylabel("Loan Amount")
plt.show()
```

### As income increases, people tend to apply for higher loan amounts, but approvals cluster in the middle-income range with moderate loans.

```python
# 10. Correlation Heatmap

# Only keep numeric columns for correlation
corr = df.select_dtypes(include=['int64', 'float64']).corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Applicant income and loan amount show a weak positive correlation. Credit history has the strongest impact on loan approval.

```python
# 14. Model Building X,y 
X = df.drop("Loan_ID",axis=1)
X
```

```python
y = df.Loan_Status
```

```python
# 15. Train_test_split the data 
from sklearn.model_selection import train_test_split
```

```python
# 16. Test_size=0.2
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2 ,random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

## 17. Model Training with LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
log_reg.score(X_train,y_train)
```

```python
log_reg.score(X_test,y_test)
```

```python
y_pred_log = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred_log))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred_log)
```

```python
cm
```

```python
plt.figure(figsize = (16,4))
sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
